```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Birthweight.csv')
data.head()
```

### 1. Виділити один зі стовпців (на вибір) з файлу як об’єкт Series. Задати назви індексів цього об’єкту. Виділити підмасиви за допомогою прямої та непрямої індексацій.

```python
sub_data = data['Length']
display(sub_data.head())

#Пряма індексація:
print(sub_data.loc[2])

sub_data.index.name = 'Counter'

#Вибиваємо рядок щоб таблична індексація збилася
sub_data = sub_data.iloc[1:]
display(sub_data.head())

#Непряма індекскація:
print(sub_data.iloc[2])
```

### 2. До об’єкту DataFrame, в який записано вміст файлу, додати новий стовпець, що є результатом операцій над іншими стовпцями.

```python
data['L/Bw'] = data['Length'] / data['Birthweight']
data.head()
```

### 3. Створити ієрархічну індексацію (не менше двох рівнів) для цього об’єкту DataFrame. Продемонструвати виділення підмасивів за певною ознакою з використанням мультиіндексації. Продемонструвати розрахунок результату операції агрегування з використанням ієрархічних рівнів.

```python
#14.5 16.7

sub_data = data['L/Bw']

indexes = []
for item in sub_data:
    if item <= 14.5:
        indexes.append('Low')
    elif (item > 14.5) and (item <= 16.7):
        indexes.append('Medium')
    else:
        indexes.append('High')

sub_data.index = [sub_data.index, indexes]
sub_data.index.names = ['Counter', 'Proportion']
display(sub_data.sample(5))

#Мультиіндексація
print(sub_data[:, 'Medium'])

#Агрегація з використанням ієрархічних рівнів
sub_data = sub_data.groupby("Proportion").mean()
display(sub_data)
```

### 4. Створити декілька власних об’єктів DataFrame за такою ж тематикою, що й файл. Наприклад, якщо тема файлу – жаби, можна створити об’єкти, що містять розміри жаб, вагу, стать, кількість особин в популяції і т.д. Використати описані в теоретичних відомостях параметри методу merge для різних видів злиття даних цих об’єктів.

```python
data1 = pd.DataFrame({
    'Name': ['Samson', 'Euclid', 'Magomed', 'Charlz'],
    'Height': [54, 56, 55, 54],
    'Weight': [4.10, 4.22, 4.07, 3.99]
})

data2 = pd.DataFrame(
    {
        'Name': ['Hermes', 'Gregory', 'Verando', 'Lucrezio'],
        'Height': [57, 51, 52, 54],
        'Weight': [4.21, 4.05, 3.95, 3.89]
    },
    index=[1, 3, 4, 5])

display(pd.merge(data1, data2, how='left', left_index=True, right_index=True, suffixes=[" A"," B"]))
display(pd.merge(data1, data2, how='outer', left_index=True, right_index=True))
display(pd.merge(data1, data2, how='right', left_index=True, right_index=True))
display(pd.merge(data1, data2, how='outer', on="Name"))
display(pd.merge(data1, data2, how='left'))
display(pd.merge(data1, data2, how='outer'))
display(pd.merge(data1, data2, how='right'))
```
